## Charge It or Crash It: Predicting Credit Card Defaults with Machine Learning

Can payment history predict who will default next month?

The most significant problem for consumer finance is determining which consumers are likely to default on their payments for their credit cards. Financial institutions incur enormous losses each year due to sudden defaults, even risking their clients' trust in case they are too strict with credit conditions. Imagine using machine learning algorithms to determine whether or not a consumer will default before it happens so that measures can be taken in advance.

With the dataset(from homework 5) on the credit card consumers based in Taiwan, where 30,000 credit card clients were used to develop a model for predicting credit card default payments.

## The Problem
Each month, the credit card company must make a yes or no decision about whether a particular client is going to default or not. The data I have analyzed comprised six months of monthly payments made by individual clients, their total credit limit values ($10,000-$500,000 TWD) as well as bills and payments in each month, along with the delay in making the payment. The first problem that I faced was the extremely imbalanced classes of observations. According to the histogram below, only 22% of the observations were cases of default. It implies that the accuracy of a completely random classifier which just predicts 'not default' for all cases would be 78%. Obviously, such an approach does not make any sense as it catches nothing at all.

Such class imbalance dictated the way how I designed the experiment. I could not rely on the traditional accuracy criterion since it would give me incorrect results. Instead, I decided to use macro F1 score.

## EDA

The dataset has 30,000 credit card customers based in Taiwan who have been tracked for six months. It has a total of 24 variables with demographic data such as age, sex, marital status, and education level, among others such as credit limit and payment details.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split

df = pd.read_csv("data/UCI_Credit_Card.csv")
train_df, test_df = train_test_split(df, test_size=0.3, random_state=123)
```

### Class Imbalance

One of the initial steps was checking the distribution of the target feature. As illustrated by the chart (Figure 1) presented below, only 22% of customers were defaulters, whereas 78% had no defaults. Therefore, a simplistic model that makes predictions based on the absence of a default will achieve 78% accuracy, which is not very useful. For this reason, I decided to use the macro-average F1 score metric for performance evaluation.

```python
plt.figure(figsize=(8, 5))
default_counts = train_df["default.payment.next.month"].value_counts()
plt.bar(['No Default', 'Default'], default_counts.values, color=['green', 'red'], alpha=0.7)
plt.title("Figure 1: Class Distribution: Default vs No Default")
plt.ylabel("Number of Customers")
for i, v in enumerate(default_counts.values):
    plt.text(i, v + 1000, f"{v} ({v/len(train_df)*100:.1f}%)", ha='center')
plt.show()
```

### Fetaure Scale Differences

The first difference that struck me about these features is the **huge disparity in scale** between them. Codes representing payment status, which are PAY_0 to PAY_6, vary between around -2 and 8, whereas bill amounts, BILL_AMT1-BILL_AMT6, have values up into the hundreds of thousands of TWD. See Figure 2 for a visual representation of this, using side-by-side box plots of both kinds of features; payment status appears very concentrated, whereas bill/payment amounts are highly dispersed and have many outliers.

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
train_df[['PAY_0', 'PAY_2', 'PAY_3']].boxplot(ax=axes[0])
axes[0].set_title("Payment Status Features (Small Scale)")
axes[0].set_ylabel("Value")
train_df[['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3']].boxplot(ax=axes[1])
axes[1].set_title("Bill Amount Features (Large Scale)")
axes[1].set_ylabel("Value")
train_df[['PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3']].boxplot(ax=axes[2])
axes[2].set_title("Payment Amount Features (Large Scale)")
axes[2].set_ylabel("Value")
plt.suptitle("Figure 2: Feature Scales Across Feature Groups", y=1.02)
plt.tight_layout()
plt.show()
```

### Correlation Between Late Payment and Bill Amount

The heatmap in Figure 3 represents the correlation between the six variables PAY_0 to PAY_6 related to monthly late payments and six variables representing monthly bill amounts, BILL_AMT1 to BILL_AMT6. The fact that there is little correlation between the number of days a person has delayed payment and his/her monthly bill amount is surprising, as high bill amount does not mean late payments.

```python
payment_delays = ['PAY_0', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
bill_amounts = ['BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6']
corr_data = train_df[payment_delays + bill_amounts].corr()
payment_bill_corr = corr_data.loc[payment_delays, bill_amounts]

plt.figure(figsize=(10, 6))
sns.heatmap(payment_bill_corr, annot=True, cmap='coolwarm', center=0,
            cbar_kws={'label': 'Correlation'}, fmt='.2f')
plt.title("Figure 3: Correlation Between Payment Delays and Bill Amounts")
plt.xlabel("Bill Amount Month")
plt.ylabel("Payment Status Month")
plt.tight_layout()
plt.show()
```

## Cavets 
1. Lack of demographic information: The input dataset consists only of financial transactions data, with no demographic attributes such as employment status or income stability that could influence customer default risk. A customer with an exemplary track record can become suddenly unemployed and therefore be at very high risk of defaulting next month, whereas our model would never realize that.

2. Model bias due to temporal drift: These data are from 2005 Taiwan; consumer behavior and the economic conditions have certainly evolved since then, so that the model has learned payment patterns before the financial crisis. The definition of "risk" behavior may have shifted since 2005 Taiwan and may certainly be different in 2024 or for any other country.

3. Class imbalance is still an issue: We used class weighting and the macro F1 score, yet we had 3.5 times more non-defaulters in the training dataset than defaulter instances, meaning there were much more "safe customer" samples to learn from than "risky customer" samples. Given that we have only 6,600 defaulter samples, some rare risk factors may be missed.


## Concluding Remarks

Despite the machine learning solution being clearly superior to making wild guesses, with an F1 score of 0.695 versus a baseline of roughly 0.50, the success was not achieved simply by choosing a better algorithm. Rather, it required taking into account the peculiarities of the dataset itself.

Some aspects of the data that emerged during EDA would affect all subsequent decisions. First of all, there are unknown values in the EDUCATION column (5 and 6) and PAY_* columns (-2 and 0) that are not defined anywhere in the original paper. Given that only 23 features are available and the BILL_AMT and PAY_AMT groups have very high correlation between themselves and the months, the feature space was quite limited, which explains the introduction of new features such as total bill amount, payment proportion, average delay, and credit utilization.

As for the choice of the model, the logistic regression algorithm demonstrated significant underfitting, proving the nonlinearity of relations in this data. The light gradient boosting machine proved to be the best choice for our problem, and the difference between its cross-validation score (0.707) and test score (0.695) suggests that no significant optimization bias is present.

All in all, however, I would suggest utilizing the model simply for preliminary screening purposes rather than making decisions based purely on its predictions. At a 35% rate of false negatives, there are likely to be some number of people who have defaulted, but who have not been flagged by the algorithm.
